```python
import pandas as pd
import os
from helper import create_folder
```

```python
def extract_players_data(folder_path):
    # Create an empty list to store player dataframes
    player_dfs = []

    # Iterate through the subdirectories in the main folder
    for root, dirs, files in os.walk(folder_path):
        for dir in dirs:
            match_uid_folder_path = os.path.join(root, dir)

            # Iterate through the files in the match_uid folder
            for file in os.listdir(match_uid_folder_path):
                if file.endswith('_outfield.csv'):  # Check if file ends with '_outfield.csv'
                    file_path = os.path.join(match_uid_folder_path, file)

                    # Read the CSV file into a DataFrame
                    df = pd.read_csv(file_path)

                    # Extract the player UID and player name columns
                    player_uids = df['player_uid'].unique()
                    player_names = df['Player'].unique()

                    # Create a DataFrame with player UID and player name
                    player_data = pd.DataFrame({'player_uid': player_uids, 'player_name': player_names})

                    # Append the player data to the list
                    player_dfs.append(player_data)

    # Concatenate the player dataframes into a single DataFrame
    player_df = pd.concat(player_dfs, ignore_index=True)

    # Remove duplicate rows from the DataFrame
    player_df = player_df.drop_duplicates().reset_index(drop=True)

    return player_df
```

```python
# Call the function and save the output to CSV
player_folder_path = 'player_data'
create_folder(player_folder_path)

players_df = extract_players_data("match_data")

players_df.to_csv(f'{player_folder_path}/players_epl_22_23.csv', index=False)
```
